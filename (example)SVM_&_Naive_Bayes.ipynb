## An example of SVM, Naive Bayes binary clasification methods on the Iris dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

```python
iris = datasets.load_iris()
iris
```

```python
x = iris.data[:,:2]
y = iris.target
x
```

```python
from sklearn.svm import SVC

svc = SVC(kernel='linear', C=1 ,gamma=1)
svc.fit(x,y)
```

```python
x_min, x_max = x[:,0].min()-1, x[:,0].max()+1
y_min, y_max = x[:,1].min()-1, x[:,1].max()+1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
plt.figure(figsize=[14, 9])
z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)
plt.contourf(xx, yy, z, alpha=0.8)
plt.scatter(x[:, 0], x[:, 1], c=y, )
plt.xlabel('Sepel lenght')
plt.ylabel('Sepel width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear karnel')
plt.show()
```

```python
svc = SVC(kernel='rbf', C=1 ,gamma=1).fit(x,y)
x_min, x_max = x[:,0].min()-1, x[:,0].max()+1
y_min, y_max = x[:,1].min()-1, x[:,1].max()+1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
plt.figure(figsize=[14, 9])
z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)
plt.contourf(xx, yy, z, alpha=0.8)
plt.scatter(x[:, 0], x[:, 1], c=y, )
plt.xlabel('Sepel lenght')
plt.ylabel('Sepel width')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with rbf karnel')
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB().fit(iris.data, iris.target)
y_pred = gnb.predict(iris.data)

print(iris.data.shape[0],(iris.target != y_pred).sum())
```

```python
from sklearn.metrics import accuracy_score

acc_gnb = accuracy_score(y_pred, iris.target)
acc_gnb
```
